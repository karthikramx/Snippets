```python
import pandas as pd

data = pd.read_csv('data/pokemon_data.csv')
#read from excel files
data_excel = pd.read_excel('data/pokemon_data.xlsx')
#readind from text files
data_text = pd.read_csv('data/pokemon_data.txt', delimiter='\t')
```

```python
# print all the columns
data.columns
```

```python
# get values of a single column
# can also be written as --? data.Name and data['Name']
data['Name']
```

```python
# Get multiple columns
data[['Name','Attack','Speed']]
```

```python
# Get first 5 rows of multiple columns 
data[['Name','Attack','Speed']][0:5]
```

```python
# integer location based extraction - iloc : integer location
data.iloc[0:4]
```

```python
# extracting value of a specific cell is down as follows - to get value of 0th row and 1st column
data.iloc[0,1]
```

```python
# iterating through rows
for index, row in data.iterrows():
    print(index, row[['Name']])
```

```python
# locating information bases on textual information
data.loc[data['Type 1'] == 'Fire']
```

```python
# get basic stats on data frame
data.describe()
```

```python
# sorting values in a data frame
data.sort_values(['Name']).head(5)
```

```python
# sorting values by multiple columns
data.sort_values(['Name','Defense'], ascending=False).head(5)
```

```python
# sorting using ascendiing mask
data.sort_values(['Name','HP'], ascending=[1,0]).head(5)
```

```python
# making changes to the tqble, adding new columns
data['Total'] = data['Attack'] + data['Defense']
```

```python
data.head(5)
```

```python
# droping a specific column from a data frame
data.drop('Total', inplace=True, axis=1)
```

```python
data.head(5)
```

```python
# performing row operations using iloc
data['Total'] = data.iloc[:,4:10].sum(axis=1)
```

```python
data.head(5)
```

```python
# Rearrranging columns in a data frame
cols = list(data.columns)
data = data[cols[0:4] + [cols[-1]] + cols[5:12]]
```

```python
data.head(5)
```

```python
# saving back as a csv file
data.to_csv('data/modified_pokemon_data.csv', index = False)
data.to_excel('data/modified_pokemon_data.xlsx',index = False)
data.to_csv('data/modified_pokemon_data.txt',index = False, sep = '\t')
```

#### Filtering Data

```python
# use & instead of and
new_df = data.loc[(data['Type 1'] == 'Grass') & (data['Type 2'] == 'Poison')]
new_df.head(5)
```

#### Reset Index

```python
new_df.reset_index(drop = True, inplace=True)
```

```python
new_df.head(10)
```

```python
# extracting names the contain a certain text
data[~data['Name'].str.contains('Mega')].head(5)
```

#### Passing regular expressions


```python
import re
```

```python
data.loc[ data['Type 1'].str.contains('file|grass', flags=re.I, regex=True) ]
# re.I refers to Ignore
```

```python
# stating with a centain alphabets
data.loc[ data['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True) ]
```

#### Conditional Statements


```python
data.loc[data['Type 1'] == 'Fire', 'Type 1' ] = 'Flamer'
data.head(10)
```

```python
data.loc[ data['Total'] > 500, ['Legendary','Generation']] = 'TEST VALUE'
```
